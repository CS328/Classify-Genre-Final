```python
import os
import sys
import math

import matplotlib
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np
import pandas as pd

from scipy.io.wavfile import read
from scipy.fft import fft, ifft
import pickle


matplotlib.style.use('ggplot')
%matplotlib notebook
```

```python

data = read("speaker-data-enter_sandman-2.wav")
audio = data[1]

window = audio[200000: 300000]

plt.figure(figsize=(10,5))
plt.plot(window)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.title("Enter Sandman")
plt.show()

plt.figure(figsize=(10,5))
windowfft = fft(window)
plt.plot(abs(windowfft[: , 1]))
plt.title("Enter Sandman FFT")
plt.show()
```
